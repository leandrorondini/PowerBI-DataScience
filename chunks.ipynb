```python
# Célula 10 - Criar chunks organizados
def criar_chunks_inteligentes(conteudo):
    chunks = []

    # Dividir por páginas primeiro
    paginas = conteudo.split('--- PÁGINA')

    for pagina in paginas[1:]:  # Pular o primeiro item vazio
        linhas = pagina.split('\n')
        numero_pagina = linhas[0].replace('---', '').strip()

        chunk_atual = f"Página {numero_pagina}:\n"

        for linha in linhas[1:]:
            linha = linha.strip()
            if linha and len(linha) > 10:
                # Se a linha parece ser um novo tópico
                if (linha.isupper() or
                    'Figure' in linha or
                    'Table' in linha or
                    len(chunk_atual) > 400):  # Limite de tamanho

                    if len(chunk_atual) > 50:
                        chunks.append(chunk_atual.strip())
                    chunk_atual = f"Página {numero_pagina}:\n"

                chunk_atual += linha + "\n"

        if len(chunk_atual) > 50:
            chunks.append(chunk_atual.strip())

    return chunks

# Aplicar nos dados extraídos
chunks = criar_chunks_inteligentes(conteudo_estrategico)
print(f"📦 Criados {len(chunks)} chunks:")
for i, chunk in enumerate(chunks[:3]):
    print(f"\n--- CHUNK {i+1} ---")
    print(chunk[:300] + "...")
```

```python
# Célula 11 - Gerar embeddings
!pip install sentence-transformers

from sentence_transformers import SentenceTransformer
import numpy as np

# Carregar modelo leve
model = SentenceTransformer('all-MiniLM-L6-v2')

print("🔢 Criando embeddings para os chunks...")
embeddings = model.encode(chunks)

print(f"✅ Embeddings criados: {len(embeddings)} chunks")
print(f"📐 Dimensões: {embeddings[0].shape}")

# Preparar dados para exportação
dados_exportar = []
for i, (chunk, embedding) in enumerate(zip(chunks, embeddings)):
    dados_exportar.append({
        'id': i,
        'chunk': chunk,
        'embedding': embedding.tolist()
    })
```

```python
# Célula 12 - Salvar dados completos para usar no OptVet
import json
import pandas as pd

# Preparar dados para exportação
dados_completos = []
for i, (chunk, embedding) in enumerate(zip(chunks, embeddings)):
    dados_completos.append({
        'id': i,
        'chunk_text': chunk,
        'embedding': embedding.tolist(),
        'tokens_count': len(chunk.split()),
        'source': 'livro_estrategico'
    })

# Salvar como JSON
with open('optvet_knowledge_base.json', 'w', encoding='utf-8') as f:
    json.dump(dados_completos, f, ensure_ascii=False, indent=2)

# Salvar como CSV (mais simples para o Base44)
df = pd.DataFrame(dados_completos)
# Não salvar embeddings no CSV (muito grande)
df_simple = df[['id', 'chunk_text', 'tokens_count', 'source']]
df_simple.to_csv('optvet_chunks_simple.csv', index=False, encoding='utf-8')

print("💾 Dados salvos:")
print(f"📁 JSON completo: {len(dados_completos)} chunks")
print(f"📊 CSV simplificado: {len(df_simple)} chunks")

# Fazer download
files.download('optvet_knowledge_base.json')
files.download('optvet_chunks_simple.csv')
```

```python
# Célula 13 - Testar sistema de busca
def buscar_conteudo_semantico(consulta, chunks, embeddings, top_k=5):
    """Buscar chunks mais similares à consulta"""
    # Embedding da consulta
    consulta_embedding = model.encode([consulta])[0]

    # Calcular similaridades
    similaridades = []
    for i, embedding in enumerate(embeddings):
        simil = np.dot(consulta_embedding, embedding) / (
            np.linalg.norm(consulta_embedding) * np.linalg.norm(embedding)
        )
        similaridades.append((simil, i))

    # Ordenar por similaridade
    similaridades.sort(reverse=True)

    # Retornar top resultados
    resultados = []
    for simil, idx in similaridades[:top_k]:
        resultados.append({
            'similarity': float(simil),
            'chunk': chunks[idx],
            'chunk_id': idx
        })

    return resultados

# Testar com algumas consultas
consultas_teste = [
    "conjunctivitis in dogs",
    "corneal ulcer treatment",
    "glaucoma symptoms",
    "eye emergency in cats"
]

print("🔍 TESTANDO BUSCA SEMÂNTICA:")
print("=" * 50)

for consulta in consultas_teste:
    print(f"\n🔎 Consulta: '{consulta}'")
    resultados = buscar_conteudo_semantico(consulta, chunks, embeddings, top_k=3)

    for i, resultado in enumerate(resultados):
        print(f"   {i+1}. Similaridade: {resultado['similarity']:.3f}")
        print(f"      Chunk: {resultado['chunk'][:100]}...")
```

```python
# Célula 15 - Refinar chunks para remover seções vazias
def refinar_chunks(chunks):
    chunks_refinados = []

    for chunk in chunks:
        # Remover chunks que são apenas títulos/seções
        linhas = chunk.split('\n')
        linhas_conteudo = [linha for linha in linhas
                          if len(linha.strip()) > 30 and  # Linhas mais substanciais
                          not linha.strip().endswith('...') and
                          'Section' not in linha and
                          'Foundation' not in linha]

        if len(linhas_conteudo) >= 2:  # Pelo menos 2 linhas de conteúdo
            chunk_refinado = '\n'.join(linhas_conteudo)
            if len(chunk_refinado) > 100:  # Chunks com conteúdo real
                chunks_refinados.append(chunk_refinado)

    return chunks_refinados

print("🔄 Refinando chunks...")
chunks_refinados = refinar_chunks(chunks)
print(f"📦 Chunks antes: {len(chunks)} | depois: {len(chunks_refinados)}")

# Recriar embeddings apenas para chunks refinados
embeddings_refinados = model.encode(chunks_refinados)
print(f"🔢 Novos embeddings: {len(embeddings_refinados)} chunks")
```

```python
# Célula 16 - Teste com chunks refinados
print("🔍 TESTE COM CHUNKS REFINADOS:")
print("=" * 50)

for consulta in consultas_teste:
    print(f"\n🔎 Consulta: '{consulta}'")
    resultados = buscar_conteudo_semantico(consulta, chunks_refinados, embeddings_refinados, top_k=3)

    for i, resultado in enumerate(resultados):
        print(f"   {i+1}. Similaridade: {resultado['similarity']:.3f}")
        print(f"      {resultado['chunk'][:120]}...")
```
